```python
def tendency_check(df):
    
    col_x = df[df.columns[0]]
    col_y = df[df.columns[1]]
    
    slope, intercept, r_value, p_value, std_err, = linregress(col_x, col_y)
    
    
    fig, ax = plt.subplots(figsize=(14, 6))
    fig.suptitle(f"Scatter {df.columns[0]} | {df.columns[1]}", fontsize=20)
    sns.set_style("darkgrid")
    
    
    ax = sns.scatterplot(
        x=col_x, 
        y=col_y,
        color=colors[5],
        alpha=1,
        s=100
    )
    ax = sns.lineplot(
        x=col_x, 
        y=col_x*slope,
        color=colors[7], 
        lw=3
    )
    ax = sns.lineplot(
        x=col_x, 
        y=col_x*slope + intercept,
        color=colors[9],
        lw=3
    )
    
    ax.set_xlabel(df.columns[0], loc="center", fontsize=15)
    ax.set_ylabel(df.columns[1], loc="center", fontsize=15)#, c=colors[color1])
    fig.legend([ "slope alone","slope with intercept", "data"],loc=(0.83, 0.75))
    
    return fig
```

```python
def residual_check(df):
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))
    fig.suptitle("Check of non linearity in the data", fontsize=20)
    
    cols = df.columns
    
    x_col = cols[0]
    y_col = cols[1]
    slope, intercept, r_value, p_value, std_err = linregress(df[x_col], df[y_col])
    
    y_pred = intercept + slope * df[x_col]    
    residuals = df[y_col] - y_pred
    lowess = sm.nonparametric.lowess(residuals, df[x_col], frac=0.5)
    
    
    
    sns.scatterplot(
        ax=ax[0],
        x=x_col,
        y=y_col,
        data=df,
        color=colors[0],
        alpha=1,
        s=100,
        label="scatter distribution"
    )
    
    sns.lineplot(
        ax=ax[0],
        x=df[x_col], 
        y=intercept + slope * df[x_col],
        color=colors[2], 
        lw=2,
        label="trendline"
        )
    
    ax[0].set_xlabel(x_col)
    ax[0].set_ylabel(y_col)
    ax[0].text(df[x_col].min(), df[y_col].max(), f"r={r_value:.2f}, p={p_value:.2f}")
    ax[0].set_title("Linear regression")
    
    # scattter !!
    sns.scatterplot(
        ax=ax[1],
        y=residuals,
        x=x_col,
        data=df,
        color=colors[4],
        alpha=1,
        label="residual"
    )
    sns.lineplot(
        ax=ax[1],
        y=lowess[:,1],
        x=lowess[:,0],
        color=colors[6],
        alpha=1,
        lw=3,
        label="smooth residual"
    )
    
    ax[1].set_xlabel("Residuals")
    ax[1].set_ylabel("Frequency")
    ax[1].set_title("Residual linearity check")
    fig.show()
```
